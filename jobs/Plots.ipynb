```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
### Seaborn style
sns.set_style("whitegrid")
```

```python
### Let's import our data
raw_data = pd.read_csv('../data/311.csv')
### and test if everything OK
raw_data.head().T
```

```python
raw_data['Agency'].value_counts()
```

```python
raw_data.plot(kind='scatter', x="Longitude", y='Latitude')
```

```python
raw_data["Created Date"] = pd.to_datetime(raw_data["Created Date"])
```

```python
raw_data["date_created"] = raw_data["Created Date"].dt.date
cnt_srs = raw_data['date_created'].value_counts()
print(cnt_srs)

plt.figure()
ax = plt.subplot(111)
ax.bar(cnt_srs.index, cnt_srs.values, alpha=0.8)
ax.xaxis_date()
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
#ax.xaxis.set_major_locator(mdates.MonthLocator(2)) 
plt.xticks(rotation='vertical')
plt.show()
```
